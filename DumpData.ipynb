```python
# Split the 'budget' column into 'budget_min' and 'budget_max' columns
df[['budget_min', 'budget_max']] = df['new_budget_3'].str.split(' - ', expand=True)

# Handle the special cases of '0-0', '0-115 or 120', and '0-500'
df.loc[df['budget_min'] == '0', ['budget_min', 'budget_max']] = [0, 0]
df.loc[df['budget_min'] == '0-115 or 120', ['budget_min', 'budget_max']] = [0, 120]
df.loc[df['budget_min'] == '0-500', ['budget_min', 'budget_max']] = [0, 500]

# Convert the 'budget_min' and 'budget_max' columns to numeric types
df['budget_min'] = pd.to_numeric(df['budget_min'], errors='coerce')
df['budget_max'] = pd.to_numeric(df['budget_max'], errors='coerce')

# Replace missing values in 'budget_min' and 'budget_max' with the mean value of the column
budget_mean = df[['budget_min', 'budget_max']].mean()
df[['budget_min', 'budget_max']] = df[['budget_min', 'budget_max']].fillna(value=budget_mean)

# Print the updated DataFrame
df
```

```python
df['budget_mean'] = (df['budget_min'] + df['budget_max']) / 2
```
